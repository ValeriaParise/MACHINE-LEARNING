```python
# Caricamento dataset
titanic = sns.load_dataset("titanic")
print("Prime righe del dataset Titanic:")
titanic.head()
```

```python
# Informazioni generali
print("\nInfo:")
print(titanic.info())
```

```python
# Statistiche
print("\nStatistiche numeriche:")
titanic.describe()
```

```python
# Controllo valori mancanti
print("\nValori mancanti per colonna:")
print(titanic.isna().sum())
```

```python
# Sostituzione valori mancanti per alcune colonne
titanic["age"] = titanic["age"].fillna(titanic["age"].median())  # et√† ‚Üí mediana
titanic["embarked"] = titanic["embarked"].fillna(titanic["embarked"].mode()[0])  # porto ‚Üí moda
```

```python
# Eliminazione colonne con troppi NaN
titanic = titanic.drop(columns=["deck"])
print("\nValori mancanti dopo la pulizia:")
print(titanic.isna().sum())
```

```python
titanic.head()
```

```python
# Eliminazione colonne con informazioni ridondanti o non utili
titanic.drop(["class", "alive", "embark_town", "who"], axis=1, inplace=True)
```

```python
# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(data=titanic, x="fare")
plt.title("Distribuzione del prezzo del biglietto (outlier inclusi)")
plt.show()
```

```python
# Rilevamento outlier tramite IQR (InterQuartile Range)
Q1 = titanic["fare"].quantile(0.25)
Q3 = titanic["fare"].quantile(0.75)
IQR = Q3 - Q1
# Definiamo i limiti per gli outlier
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtriamo i valori validi
titanic_clean = titanic[(titanic["fare"] >= lower_limit) & (titanic["fare"] <= upper_limit)]

print(f"Righe originali: {len(titanic)}, dopo rimozione outlier: {len(titanic_clean)}")
```

```python
# Creazione di una nuova feature: dimensione della famiglia 
titanic_clean["family_size"] = titanic_clean["sibsp"] + titanic_clean["parch"] + 1
```

```python
# Trasformazione di variabili categoriche
le = LabelEncoder()
titanic_clean["sex"] = le.fit_transform(titanic_clean["sex"])
titanic_clean.head()
```

```python
titanic_clean["embarked"] = le.fit_transform(titanic_clean["embarked"]) #potremmo farlo pure con one-hot encoding
titanic_clean.head()
```

```python
# Scaling 
scaler = MinMaxScaler()
titanic_clean[["age", "fare"]] = scaler.fit_transform(titanic_clean[["age", "fare"]])
print(titanic_clean.head())
```

```python
sns.countplot(data=titanic_clean, x="survived")
```

```python
# SMOTE (Synthetic Minority Oversampling Technique)
# Bilancia le classi creando nuovi campioni sintetici

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(titanic_clean.drop(columns=["survived"]), titanic_clean["survived"])

print("\nDistribuzione classi originale:")
print(titanic_clean["survived"].value_counts())

print("\nDistribuzione classi dopo SMOTE:")
print(pd.Series(y_res).value_counts())

titanic_res = pd.concat([pd.DataFrame(X_res, columns=X_res.columns), pd.Series(y_res, name="survived")], axis=1)
sns.countplot(data=titanic_res, x="survived")
```

```python
# Rilevamento outlier tramite IQR (InterQuartile Range)
Q1 = tips["total_bill"].quantile(0.25)
Q3 = tips["total_bill"].quantile(0.75)
IQR = Q3 - Q1
# Definiamo i limiti per gli outlier
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtriamo i valori validi
tips_clean = tips[(tips["total_bill"] >= lower_limit) & (tips["total_bill"] <= upper_limit)]

print(f"Righe originali: {len(tips)}, dopo rimozione outlier: {len(tips_clean)}")
```

```python
# Nuova feature: rapporto mancia/conto
tips_clean["tip_to_bill_ratio"] = tips_clean["tip"] / tips_clean["total_bill"]
tips_clean.head()
```
